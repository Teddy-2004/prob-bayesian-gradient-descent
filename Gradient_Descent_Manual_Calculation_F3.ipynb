# Part 3: Gradient Descent Manual Calculation

This notebook manually computes three iterations of gradient descent with:
- **Full chain rule derivation** (step-by-step)
- **Explicit arithmetic** (no compressed steps)
- **Deep reflection** (conceptual understanding)

## SETUP: PROBLEM DEFINITION

```python
import numpy as np
import pandas as pd

# Initial parameters
m = -1
b = 1
alpha = 0.1  # learning rate

# Data points
data_points = [(1, 3), (3, 6)]
X = np.array([point[0] for point in data_points])
y = np.array([point[1] for point in data_points])
n = len(data_points)

print(f'Initial: m = {m}, b = {b}')
print(f'Learning rate: α = {alpha}')
print(f'Data points: {data_points}')
```

## SECTION 1: FULL CHAIN RULE DERIVATION

This section shows the COMPLETE derivative calculation.

```python
print('FULL CHAIN RULE DERIVATION FOR ∂J/∂m')

print('\n1. COST FUNCTION:')
print('   J(m,b) = (1/n) Σ(yᵢ - (mxᵢ + b))²')
print('   J(m,b) = (1/n) Σ(yᵢ - mxᵢ - b)²')

print('\n2. CHAIN RULE APPLICATION:')
print('   We need: ∂J/∂m')
print('   ')
print('   By chain rule: ∂J/∂m = (1/n) Σ ∂/∂m[(yᵢ - mxᵢ - b)²]')

print('\n3. OUTER DERIVATIVE (apply power rule):' )
print('   ∂/∂m[(yᵢ - mxᵢ - b)²] = 2(yᵢ - mxᵢ - b) × ∂/∂m[yᵢ - mxᵢ - b]')
print('                           ^                    ^')
print('                     [Outer]              [Inner chain]')

print('\n4. INNER DERIVATIVE (derivative of the argument):' )
print('   ∂/∂m[yᵢ - mxᵢ - b] = 0 - xᵢ - 0')
print('                       = -xᵢ')
print('   (Since yᵢ and b are constants w.r.t. m)')

print('\n5. MULTIPLY OUTER AND INNER:')
print('   ∂/∂m[(yᵢ - mxᵢ - b)²] = 2(yᵢ - mxᵢ - b) × (-xᵢ)')
print('                          = -2(yᵢ - mxᵢ - b)xᵢ')

print('\n6. REWRITE IN STANDARD FORM:')
print('   Note: (yᵢ - mxᵢ - b) = -(mxᵢ + b - yᵢ)')
print('   So: -2(yᵢ - mxᵢ - b)xᵢ = -2[-(mxᵢ + b - yᵢ)]xᵢ')
print('                           = 2(mxᵢ + b - yᵢ)xᵢ')

print('\n7. SUM OVER ALL DATA POINTS AND AVERAGE:')
print('   ∂J/∂m = (1/n) Σ 2(mxᵢ + b - yᵢ)xᵢ')
print('         = (2/n) Σ(mxᵢ + b - yᵢ)xᵢ')

print('\nSIMILARLY FOR ∂J/∂b:')

print('\n∂J/∂b = (1/n) Σ ∂/∂b[(yᵢ - mxᵢ - b)²]')
print('      = (1/n) Σ 2(yᵢ - mxᵢ - b) × ∂/∂b[yᵢ - mxᵢ - b]')
print('      = (1/n) Σ 2(yᵢ - mxᵢ - b) × (-1)')
print('      = (1/n) Σ -2(yᵢ - mxᵢ - b)')
print('      = (2/n) Σ(mxᵢ + b - yᵢ)')
```

## ITERATION 1

```python
print('ITERATION 1: STEP-BY-STEP CALCULATION')

m_iter = -1
b_iter = 1

print(f'\nStarting: m = {m_iter}, b = {b_iter}')

# Step 1: Predictions
print('\nSTEP 1: COMPUTE PREDICTIONS ŷ = mx + b')

y_pred = []
for i, x_val in enumerate(X):
    y_hat = m_iter * x_val + b_iter
    y_pred.append(y_hat)
    print(f'\nPoint {i+1}: x = {x_val}')
    print(f'  ŷ = mx + b')
    print(f'  ŷ = ({m_iter}) × {x_val} + {b_iter}')
    print(f'  ŷ = {m_iter * x_val} + {b_iter}')
    print(f'  ŷ = {y_hat}')

y_pred = np.array(y_pred)

# Step 2: Errors
print('\nSTEP 2: COMPUTE ERRORS (ŷ - y)')

errors = []
for i, (y_hat, y_val) in enumerate(zip(y_pred, y)):
    error = y_hat - y_val
    errors.append(error)
    print(f'\nPoint {i+1}:')
    print(f'  Error = ŷ - y')
    print(f'  Error = {y_hat} - {y_val}')
    print(f'  Error = {error}')

errors = np.array(errors)

# Step 3: MSE
print('\nSTEP 3: COMPUTE MSE COST')

print('\nJ(m,b) = (1/n) Σ(ŷ - y)²')
squared_errors = []
for i, error in enumerate(errors):
    sq = error ** 2
    squared_errors.append(sq)
    print(f'\nPoint {i+1}: ({error})² = {sq}')

sum_sq = sum(squared_errors)
mse_1 = sum_sq / n

print(f'\nSum of squared errors: {squared_errors[0]} + {squared_errors[1]} = {sum_sq}')
print(f'J(m,b) = {sum_sq} / {n} = {mse_1}')

# Step 4: Gradients with EXPLICIT arithmetic
print('\nSTEP 4: COMPUTE GRADIENTS - EXPLICIT ARITHMETIC')

print('\n∂J/∂m = (2/n) Σ(mxᵢ + b - yᵢ) × xᵢ')
print('\nCompute each term (mxᵢ + b - yᵢ) × xᵢ:')

grad_m_terms = []
for i, (y_hat, y_val, x_val) in enumerate(zip(y_pred, y, X)):
    term_val = y_hat - y_val  # This is (mxᵢ + b - yᵢ)
    product = term_val * x_val
    grad_m_terms.append(product)
    print(f'\nPoint {i+1}:')
    print(f'  (mxᵢ + b - yᵢ) × xᵢ')
    print(f'  = ({y_hat} - {y_val}) × {x_val}')
    print(f'  = ({term_val}) × {x_val}')
    print(f'  = {product}')

sum_grad_m = sum(grad_m_terms)
print(f'\nSum: {grad_m_terms[0]} + {grad_m_terms[1]} = {sum_grad_m}')

grad_m_1 = (2 / n) * sum_grad_m
print(f'∂J/∂m = (2/{n}) × {sum_grad_m} = {grad_m_1}')

print('\n∂J/∂b = (2/n) Σ(mxᵢ + b - yᵢ)')
print('\nCompute sum of (mxᵢ + b - yᵢ):')

grad_b_terms = []
for i, (y_hat, y_val) in enumerate(zip(y_pred, y)):
    term_val = y_hat - y_val
    grad_b_terms.append(term_val)
    print(f'\nPoint {i+1}:')
    print(f'  (mxᵢ + b - yᵢ) = {y_hat} - {y_val} = {term_val}')

sum_grad_b = sum(grad_b_terms)
print(f'\nSum: {grad_b_terms[0]} + {grad_b_terms[1]} = {sum_grad_b}')

grad_b_1 = (2 / n) * sum_grad_b
print(f'∂J/∂b = (2/{n}) × {sum_grad_b} = {grad_b_1}')

# Step 5: Updates
print('\nSTEP 5: UPDATE PARAMETERS')

print('\nm := m - α × (∂J/∂m)')
m_new_1 = m_iter - alpha * grad_m_1
print(f'  = {m_iter} - {alpha} × {grad_m_1}')
print(f'  = {m_iter} - {alpha * grad_m_1}')
print(f'  = {m_new_1}')

print('\nb := b - α × (∂J/∂b)')
b_new_1 = b_iter - alpha * grad_b_1
print(f'  = {b_iter} - {alpha} × {grad_b_1}')
print(f'  = {b_iter} - {alpha * grad_b_1}')
print(f'  = {b_new_1}')

print('\nITERATION 1 SUMMARY')
print(f'\nBefore: m = {m_iter}, b = {b_iter}')
print(f'Cost:   J = {mse_1:.4f}')
print(f'After:  m = {m_new_1:.4f}, b = {b_new_1:.4f}')
print(f'Change: Δm = {m_new_1 - m_iter:.4f}, Δb = {b_new_1 - b_iter:.4f}')

m, b = m_new_1, b_new_1
```

## ITERATION 2: it's ABBREVIATED as full structure already shown in Iteration 1.

```python
print('\nITERATION 2\n')

# Get the updated parameters from iteration 1
m_iter = m_new_1
b_iter = b_new_1
y_pred = m_iter * X + b_iter
errors = y_pred - y
mse_2 = np.mean(errors ** 2)

grad_m_2 = (2 / n) * np.sum(errors * X)
grad_b_2 = (2 / n) * np.sum(errors)

m_new_2 = m_iter - alpha * grad_m_2
b_new_2 = b_iter - alpha * grad_b_2

print(f'Predictions: ŷ = [{m_iter:.4f} × 1 + {b_iter:.4f}, {m_iter:.4f} × 3 + {b_iter:.4f}]')
print(f'            ŷ = [{y_pred[0]:.4f}, {y_pred[1]:.4f}]')
print(f'Errors: [{errors[0]:.4f}, {errors[1]:.4f}]')
print(f'Cost: J = {mse_2:.4f}')
print(f'Gradients: ∂J/∂m = {grad_m_2:.4f}, ∂J/∂b = {grad_b_2:.4f}')
print(f'Updates: m := {m_new_2:.4f}, b := {b_new_2:.4f}')

m, b = m_new_2, b_new_2
```

## ITERATION 3

```python
print('\nITERATION 3\n')

m_iter = m_new_2
b_iter = b_new_2
y_pred = m_iter * X + b_iter
errors = y_pred - y
mse_3 = np.mean(errors ** 2)

grad_m_3 = (2 / n) * np.sum(errors * X)
grad_b_3 = (2 / n) * np.sum(errors)

m_new_3 = m_iter - alpha * grad_m_3
b_new_3 = b_iter - alpha * grad_b_3

print(f'Predictions: ŷ = [{y_pred[0]:.4f}, {y_pred[1]:.4f}]')
print(f'Errors: [{errors[0]:.4f}, {errors[1]:.4f}]')
print(f'Cost: J = {mse_3:.4f}')
print(f'Gradients: ∂J/∂m = {grad_m_3:.4f}, ∂J/∂b = {grad_b_3:.4f}')
print(f'Updates: m := {m_new_3:.4f}, b := {b_new_3:.4f}')

# After updating m and b in Iteration 3
m_new_3 = m_iter - alpha * grad_m_3
b_new_3 = b_iter - alpha * grad_b_3

# COMPUTE COST WITH NEW PARAMETERS
y_pred_new_3 = m_new_3 * X + b_new_3
errors_new_3 = y_pred_new_3 - y
mse_after_3 = np.mean(errors_new_3 ** 2)
```

## SECTION 2: COMPREHENSIVE REFLECTION

```python
print('\n' + '='*80)
print('COMPREHENSIVE REFLECTION & CONCEPTUAL UNDERSTANDING')
print('='*80)

print('\n1. TREND ANALYSIS: ERROR DECREASING MONOTONICALLY?')
print('-'*80)

initial_mse = mse_1
print(f'Initial Cost (m=-1, b=1):  J = {initial_mse:.6f}')
print(f'After Iter 1:   J = {mse_2:.6f} (decreased by {initial_mse - mse_2:.6f})')
print(f'After Iter 2:   J = {mse_3:.6f} (decreased by {mse_2 - mse_3:.6f})')
print(f'After Iter 3:   J = {mse_after_3:.6f} (decreased by {mse_3 - mse_after_3:.6f})')

print('\nYES: Error decreases MONOTONICALLY (strictly every iteration)')
print('This is GUARANTEED by gradient descent: moving opposite to gradient always')
print('reduces cost (for sufficiently small learning rate α).')

print('\n\n2. PARAMETER MOVEMENT: TOWARD LEAST SQUARES SOLUTION?')
print('-'*80)

print(f'\nSlope m trajectory:')
print(f'  Start:      m = -1.0000')
print(f'  Iter 1:     m =  1.7000 (jumped toward positive slope)')
print(f'  Iter 2:     m =  1.2600 (adjusting finer)')
print(f'  Iter 3:     m =  1.3400 (oscillating around optimum)')

print(f'\nIntercept b trajectory:')
print(f'  Start:      b = 1.0000')
print(f'  Iter 1:     b = 2.1000 (increasing toward better fit)')
print(f'  Iter 2:     b = 1.9000 (fine-tuning)')
print(f'  Iter 3:     b = 1.9160 (approaching stable value)')

print('\n** YES: Parameters are converging toward the least squares solution **')
print('The optimal solution for y = mx + b fitting (1,3) and (3,6) is approximately')
print('m ≈ 1.5, b ≈ 2.0, which our parameters are approaching.')

print('\n\n3. WHY GRADIENT DESCENT MOVES IN NEGATIVE GRADIENT DIRECTION?')
print('-'*80)

print('\nThe gradient ∇J = (∂J/∂m, ∂J/∂b) points in the direction of STEEPEST INCREASE')
print('of the cost function.')

print('\nExample from Iteration 1:')
print(f'  ∂J/∂m = -27 (NEGATIVE gradient)')
print(f'  → Cost increases when we DECREASE m')
print(f'  → So we should INCREASE m (move opposite to gradient direction)')
print(f'  → m := m - α × (-27) = m + 2.7 → m increases ✓')

print('\nThis is why we use: m := m - α × (∂J/∂m)')
print('  - If ∂J/∂m > 0: gradient positive → subtract it → m decreases')
print('  - If ∂J/∂m < 0: gradient negative → subtract it → m increases')
print('  - Result: Always move toward reducing cost')

print('\n** INSIGHT: The negative sign in "m - α∇J" ensures we climb DOWN the cost landscape **')

print('\n\n4. WHAT IF LEARNING RATE α IS TOO LARGE?')
print('-'*80)

print('\nCurrent: α = 0.1 (good balance)')
print('\nScenario A: α = 0.01 (too small)')
print('  Step size: 0.01 × 27 = 0.27 (tiny step)')
print('  Result: Very slow convergence, needs 100+ iterations')
print('  Problem: Computationally expensive')

print('\nScenario B: α = 1.0 (too large)')
print('  Step size: 1.0 × 27 = 27 (huge step)')
print('  Update: m := -1 - 27 = -28 (wildly overshoots minimum!)')
print('  Result: DIVERGENCE - cost increases instead of decreasing')
print('  Can oscillate or explode to infinity')

print('\nScenario C: α = 0.1 (our choice) - GOLDILOCKS ZONE')
print('  Step size: 0.1 × 27 = 2.7 (reasonable step)')
print('  Update: m := -1 + 2.7 = 1.7 (good progress toward optimum)')
print('  Result: Convergence in few iterations')

print('\n** CRITICAL INSIGHT: Learning rate controls the SPEED-STABILITY TRADEOFF **')
print('  - Too small: Safe but slow')
print('  - Too large: Fast but unstable/divergent')
print('  - Just right: Converges efficiently')

print('\n\n5. MATHEMATICAL GUARANTEE OF DESCENT')
print('-'*80)

print('\nTaylor expansion shows: J(m - α∇J) ≈ J(m) - α||∇J||²')
print('\nSince α > 0 and ||∇J||² ≥ 0:')
print('  → J(m - α∇J) < J(m) whenever ∇J ≠ 0')
print('  → Cost ALWAYS decreases (until gradient is zero at optimum)')

print('\nThis PROVES gradient descent works:')
print('  1. Compute gradient (direction of steepest increase)')
print('  2. Move opposite with step size α')
print('  3. Cost guaranteed to reduce')
print('  4. Repeat until convergence')

print('\n' + '='*80)
```
